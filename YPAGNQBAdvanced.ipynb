```python
def calculate_redshift(lambda_obs, lambda_emit):
    """
    Calculates the redshift (z) for a single observation.
    Formula: z = (λobs - λemit) / λemit
    """
    return (lambda_obs - lambda_emit) / lambda_emit

def main():
    # Example data: 50 aligned wavelengths (in nm)
    # In a real scenario, these could be loaded from a file or database
    obs_wavelengths = [700.0 + i for i in range(50)]
    emit_wavelengths = [656.3 for _ in range(50)]
    
    redshifts = []
    
    # Calculate individual redshifts
    for i in range(len(obs_wavelengths)):
        z = calculate_redshift(obs_wavelengths[i], emit_wavelengths[i])
        redshifts.append(z)
        
    # Calculate the average redshift
    average_redshift = sum(redshifts) / len(redshifts)
    
    # Display individual results and the summary
    print(f"{'Galaxy Index':<15} | {'Redshift (z)':<15}")
    print("-" * 35)
    for i, z in enumerate(redshifts):
        print(f"{i+1:<15} | {z:.6f}")
    
    print("-" * 35)
    print(f"Average Redshift of Sample: {average_redshift:.6f}")

if __name__ == "__main__":
    main()
```

```python
def calculate_hubble_time(h0_kms_mpc):
    """
    Calculates the Hubble time (tH) in years.
    
    Dimensional Analysis:
    1. H0 is in km/s/Mpc.
    2. Convert Mpc to km: 1 Mpc ≈ 3.086e19 km.
    3. tH = 1/H0 will be in (km/s/Mpc)^-1 * (1 Mpc / 3.086e19 km) = seconds.
    4. Convert seconds to years: 1 year ≈ 3.154e7 seconds.
    """
    # Conversion factors
    MPC_TO_KM = 3.086e19  # 1 Megaparsec in kilometers
    SEC_PER_YEAR = 3.154e7 # Seconds in a tropical year
    
    # 1. Convert H0 from (km/s/Mpc) to (1/s)
    # H0_per_s = (km/s/Mpc) / (km/Mpc) = 1/s
    h0_per_s = h0_kms_mpc / MPC_TO_KM
    
    # 2. Calculate Hubble time in seconds
    t_h_seconds = 1 / h0_per_s
    
    # 3. Convert to years
    t_h_years = t_h_seconds / SEC_PER_YEAR
    
    return t_h_years

def main():
    # Hubble Constant H0 in km/s/Mpc
    H0 = 70.0
    
    age_years = calculate_hubble_time(H0)
    age_gyr = age_years / 1e9
    
    print(f"Hubble Constant (H0): {H0} km/s/Mpc")
    print(f"Calculated Hubble Time: {age_years:} years = {age_gyr:} Gyr")

if __name__ == "__main__":
    main()
```

```python
import math

def calculate_orbital_period(semi_major_axis, total_mass):
    """
    Computes the orbital period (P) using Kepler's Third Law.
    
    Formula: P = sqrt((4 * pi^2 * a^3) / (G * M))
    
    Args:
        semi_major_axis (float): 'a' in meters.
        total_mass (float): 'M' in kilograms.
        
    Returns:
        float: Orbital period 'P' in seconds.
    """
    G = 6.67430e-11  # Gravitational constant (m^3 kg^-1 s^-2)
    
    # Calculate P^2 = (4 * pi^2 * a^3) / (G * M)
    p_squared = (4 * (math.pi**2) * (semi_major_axis**3)) / (G * total_mass)
    
    # Take the square root to get P
    return math.sqrt(p_squared)

def main():
    # Example: Calculate orbital period of Earth
    # Earth's semi-major axis 'a' ≈ 1.496e11 meters
    # Sun's mass 'M' ≈ 1.989e30 kilograms
    a = 1.496e11
    m = 1.989e30
    
    period_seconds = calculate_orbital_period(a, m)
    
    # Convert seconds to days for readability
    period_days = period_seconds / (60 * 60 * 24)
    
    print(f"Semi-major axis: {a:.2e} m")
    print(f"Total Mass:      {m:.2e} kg")
    print(f"Orbital Period:  {period_days:.4f} days")

if __name__ == "__main__":
    main()
```

```python
import math

def calculate_euclidean_distance(g1, g2):
    """Calculates 3D Euclidean distance between two galaxies."""
    dx = g1['x'] - g2['x']
    dy = g1['y'] - g2['y']
    dz = g1['z'] - g2['z']
    return math.sqrt(dx**2 + dy**2 + dz**2)

def main():
    # Simulated galaxy catalog (x, y, z in Megaparsecs)
    catalog = [
        {"name": "G1", "x": 0, "y": 0, "z": 0},
        {"name": "G2", "x": 10, "y": 5, "z": 2},
        {"name": "G3", "x": 5, "y": 12, "z": 8},
        {"name": "G4", "x": 2, "y": 8, "z": 15}
    ]
    
    n = len(catalog)
    # Initialize an empty matrix
    matrix = [[0.0 for _ in range(n)] for _ in range(n)]
    
    # Use nested loops to compute pairwise distances
    for i in range(n):
        for j in range(i + 1, n):
            dist = calculate_euclidean_distance(catalog[i], catalog[j])
            matrix[i][j] = dist
            matrix[j][i] = dist # Matrix is symmetric
            
    # Print the distance matrix
    header = "      " + "  ".join([g['name'] for g in catalog])
    print(header)
    for i, row in enumerate(matrix):
        row_str = f"{catalog[i]['name']} | " + "  ".join([f"{val:4.4f}" for val in row])
        print(row_str)

if __name__ == "__main__":
    main()
```

```python
# --- Physical Constants ---
SPEED_OF_LIGHT = 299792.458  # km/s
HUBBLE_CONSTANT = 70.0      # km/s/Mpc
EDDINGTON_CONSTANT = 1.26e38 # erg/s per solar mass

def process_agn_data(catalog):
    """Calculates physical properties for each entry."""
    for entry in catalog:
        z = entry['redshift']
        m = entry['m_bh']
        # Calculations
        entry['dist_mpc'] = (z * SPEED_OF_LIGHT) / HUBBLE_CONSTANT
        entry['edd_lum'] = EDDINGTON_CONSTANT * m
    return catalog

def main():
    # Catalog of 10 AGNs
    agn_catalog = [
        {"name": "NGC 4151", "redshift": 0.0033, "m_bh": 4.5e7},
        {"name": "3C 273",   "redshift": 0.1580, "m_bh": 8.8e8},
        {"name": "Mrk 231",  "redshift": 0.0422, "m_bh": 1.2e8},
        {"name": "NGC 1275", "redshift": 0.0176, "m_bh": 3.4e8},
        {"name": "NGC 4395", "redshift": 0.0009, "m_bh": 1.0e4},
        {"name": "Cygnus A", "redshift": 0.0561, "m_bh": 2.5e9},
        {"name": "NGC 5548", "redshift": 0.0172, "m_bh": 6.7e7},
        {"name": "NGC 3227", "redshift": 0.0039, "m_bh": 4.2e7},
        {"name": "Fairall 9", "redshift": 0.0470, "m_bh": 2.5e8},
        {"name": "NGC 7469", "redshift": 0.0163, "m_bh": 1.2e7}
    ]

    enriched_catalog = process_agn_data(agn_catalog)

    print(f"{'Name':<10} | {'z':<7} | {'Mass (M⊙)':<9} | {'Dist (Mpc)':<10} | {'Eddington (erg/s)'}")
    print("-" * 75)
    
    for entry in enriched_catalog:
        print(f"{entry['name']:<10} | {entry['redshift']:<7.4f} | "
              f"{entry['m_bh']:<10.1e} | {entry['dist_mpc']:<10.1f} | "
              f"{entry['edd_lum']:<15.4e}")

if __name__ == "__main__":
    main()
```
