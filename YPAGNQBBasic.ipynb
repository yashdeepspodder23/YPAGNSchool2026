```python
# Define the Sun's physical properties
name = "Sun"
mass_kg = 1.989e30
radius_meters = 6.9634e8

# Format the data into a readable string using f-strings
display_text = (
    f"--- Star Information ---\n"
    f"Name:   {name}\n"
    f"Mass:   {mass_kg:.2e} kg\n"
    f"Radius: {radius_meters:.2e} meters"
)

# Print the result to the console
print(display_text)
```

```python
# Constant for Sun's temperature
SUN_TEMP = 5778

# Get temperature from user and validate input
while True:
    try:
        star_temp = float(input("Enter the star's effective temperature in Kelvin (K): "))
        if star_temp < 0:
            print("Temperature cannot be negative. Please try again.")
            continue
        break
    except ValueError:
        print("Invalid input. Please enter a numeric value.")

# Determine the comparison result
if star_temp > SUN_TEMP:
    result = "hotter than"
elif star_temp < SUN_TEMP:
    result = "cooler than"
else:
    result = "the same temperature as"

# Print the result using an f-string
print(f"A star with a temperature of {star_temp} K is {result} the Sun ({SUN_TEMP} K).")
```

```python
# Define the list of planets in order of distance from the Sun
planets = [
    "Mercury", "Venus", "Earth", "Mars", 
    "Jupiter", "Saturn", "Uranus", "Neptune"
]

# Print the header
print("Planets of the Solar System (in order from the Sun):")

# Iterate through the list and print each with its sequence number
for i, planet in enumerate(planets):
    print(f"{i + 1}: {planet}")
```

```python
# List of five prominent bright stars
stars = ["Sirius", "Canopus", "Rigil Kentaurus", "Arcturus", "Vega"]

# Extract the requested positions
# Index 0 is the first, index 2 is the middle, index -1 is the last
first_star = stars[0]
middle_star = stars[2]
last_star = stars[-1]

# Print the results
print(f"First star:  {first_star}")
print(f"Middle star: {middle_star}")
print(f"Last star:   {last_star}")
```

```python
# Define the input string
target_string = "Messier87"

# Slice from index 0 up to (but not including) index 7 to get "Messier"
name = target_string[:7]

# Slice from index 7 to the end to get "87"
number = target_string[7:]

# Use step -1 to reverse the entire string
reversed_str = target_string[::-1]

# Print the results
print(f"Original string: {target_string}")
print(f"Name extracted:  {name}")
print(f"Number extracted: {number}")
print(f"Reversed string: {reversed_str}")
```
