```python
def calculate_luminosity(mass_solar_units):
    """
    Calculates luminosity using the mass-luminosity relation: L = M^3.5
    
    Args:
        mass_solar_units (float): Stellar mass in Msun.
        
    Returns:
        float: Stellar luminosity in Lsun.
    """
    # L = M^3.5
    luminosity = mass_solar_units ** 3.5
    return luminosity

def main():
    # List of stellar masses in Msun
    stellar_masses = [0.5, 1, 2, 5, 10, 20]
    
    # Define table header
    header = f"{'Mass (M⊙)':<11} | {'Luminosity (L⊙)'}"
    print(header)
    print("-" * len(header))
    
    # Calculate and print each result
    for mass in stellar_masses:
        luminosity = calculate_luminosity(mass)
        print(f"{mass:<12.1f} | {luminosity:<15.4f}")

if __name__ == "__main__":
    main()
```

```python
def calculate_wiens_peak(temperature):
    """
    Calculates the Wien's peak wavelength (lambda_max) for a given temperature.
    
    Formula: lambda_max = (2.897e-3) / T
    
    Args:
        temperature (float): Temperature in Kelvin (K).
        
    Returns:
        float: Peak wavelength in meters (m).
    """
    # Wien's displacement constant in m*K
    b = 2.897e-3
    
    return b / temperature

def main():
    # Define start, end, and step
    start_temp = 3000
    end_temp = 10000
    step = 500
    
    # Define table header
    header = f"{'Temp (K)':<10} | {'Peak Wavelength (nm)'}"
    print(header)
    print("-" * len(header))
    
    # Loop through the temperature sequence
    # range() does not include the stop value, so we use end_temp + 1
    for temp in range(start_temp, end_temp + 1, step):
        peak_m = calculate_wiens_peak(temp)
        # Convert to nanometers (nm) for readability: 1 m = 1e9 nm
        peak_nm = peak_m * 1e9
        
        print(f"{temp:<10} | {peak_nm:<20.4f}")

if __name__ == "__main__":
    main()
```

```python
def calculate_schwarzschild_radius(mass_kg):
    """Calculates Rs in meters for a given mass in kg."""
    G = 6.67430e-11
    c = 299792458
    return (2 * G * mass_kg) / (c**2)

def main():
    sun_mass_kg = 1.989e30
    
    # Generate powers of 10 from 10^5 to 10^9
    # Using range to get exponents 5, 6, 7, 8, 9
    exponents = range(5, 10)
    
    # Define table header
    header = f"{'Mass (M⊙)':<15} | {'Rs (km)':<15}"
    print(header)
    print("-" * 35)
    
    # Calculate and print table rows
    for exp in exponents:
        mass_solar = 10**exp
        mass_kg = mass_solar * sun_mass_kg
        
        rs_meters = calculate_schwarzschild_radius(mass_kg)
        rs_km = rs_meters / 1000
        
        print(f"{mass_solar:<15.0e} | {rs_km:<15.4f}")

if __name__ == "__main__":
    main()
```

```python
def calculate_gravitational_force(m1, m2, r):
    """Calculates the magnitude of gravitational force between two masses."""
    G = 6.67430e-11  # m^3 kg^-1 s^-2
    return G * (m1 * m2) / (r**2)

def main():
    # Simulated cluster: Each star contains name, mass (kg), and 1D position (meters)
    cluster = [
        {"name": "Star A", "mass": 1.989e30, "pos": 0},
        {"name": "Star B", "mass": 2.500e30, "pos": 5.0e11},
        {"name": "Star C", "mass": 1.500e30, "pos": 1.2e12}
    ]
    
    # Table header
    header = f"{'Pair':<15} | {'Mass 1 (kg)':<12} | {'Mass 2 (kg)':<12} | {'Force (N)'}"
    print(header)
    print("-" * 65)
    
    # Nested loops to iterate through unique pairs
    for i in range(len(cluster)):
        for j in range(i + 1, len(cluster)):
            star1 = cluster[i]
            star2 = cluster[j]
            
            # Distance between stars
            distance = abs(star2['pos'] - star1['pos'])
            
            # Compute force
            force = calculate_gravitational_force(star1['mass'], star2['mass'], distance)
            
            # Print row
            print(f"{star1['name']+' & '+star2['name']:<15} | "
                  f"{star1['mass']:<12.2e} | "
                  f"{star2['mass']:<12.2e} | "
                  f"{force:.4e}")

if __name__ == "__main__":
    main()
```
