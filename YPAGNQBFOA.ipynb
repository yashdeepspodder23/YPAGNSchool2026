```python
def calculate_redshift(lambda_obs, lambda_emit):
    """
    Calculates the spectroscopic redshift (z).
    
    Formula: z = (lambda_obs - lambda_emit) / lambda_emit
    
    Args:
        lambda_obs (float): Observed wavelength.
        lambda_emit (float): Rest-frame (emitted) wavelength.
        
    Returns:
        float: The calculated redshift z.
    """
    # Calculate redshift based on the provided formula
    z = (lambda_obs - lambda_emit) / lambda_emit
    return z

def main():
    # Define the observed and rest-frame wavelengths for H-alpha
    observed_wavelength = 700.0  # nm
    rest_wavelength = 656.3      # nm
    
    # Calculate the redshift using the helper function
    redshift_value = calculate_redshift(observed_wavelength, rest_wavelength)
    
    # Output the result formatted to 4 decimal places
    print(f"Observed Wavelength: {observed_wavelength} nm")
    print(f"Rest-frame Wavelength: {rest_wavelength} nm")
    print(f"Calculated Redshift (z): {redshift_value:.8f}")

if __name__ == "__main__":
    main()
```

```python
def calculate_recession_velocity(z):
    """
    Calculates the recession velocity (v) from redshift (z) 
    using the linear approximation v = c * z.
    
    Args:
        redshift (float): The spectroscopic redshift z.
        
    Returns:
        float: The recession velocity v in km/s.
    """
    # Speed of light constant in km/s
    c = 3e5
    
    # Calculate velocity
    v = c * z
    return v

def main():
    # Example redshift value (e.g., from the previous H-alpha calculation)
    # z = (700 - 656.3) / 656.3 ≈ 0.06658
    z_val = 0.06658540
    
    # Calculate velocity using the function
    v_kms = calculate_recession_velocity(z_val)
    
    # Print the result
    print(f"For a redshift (z) of {z_val:.6f}:")
    print(f"The estimated recession velocity is {v_kms:.3f} km/s.")

if __name__ == "__main__":
    main()
```

```python
def calculate_luminosity_distance(v, h0=70.0):
    """
    Calculates the luminosity distance (d) in Megaparsecs (Mpc) 
    using Hubble's Law: d = v / H0.
    
    Args:
        v (float): Recession velocity in km/s.
        h0 (float): Hubble constant in km/s/Mpc.
        
    Returns:
        float: The luminosity distance in Mpc.
    """
    # Calculate distance based on Hubble's Law
    distance = v / h0
    return distance

def main():
    # Hubble Constant (H0) in km/s/Mpc
    HUB_CONST = 70.0
    
    # Using the velocity derived from the previous H-alpha redshift example
    # (v ≈ 19974 km/s based on z ≈ 0.06658)
    rec_v = 19975.620 
    
    # Calculate distance using the helper function
    distance_mpc = calculate_luminosity_distance(rec_v, HUB_CONST)
    
    # Print the result
    print(f"Recession Velocity: {rec_v} km/s")
    print(f"Hubble Constant:    {HUB_CONST} km/s/Mpc")
    print(f"Calculated Luminosity Distance: {distance_mpc:.4f} Mpc")

if __name__ == "__main__":
    main()
```

```python
def calculate_schwarzschild_radius(mass_kg):
    """
    Calculates the Schwarzschild radius (Rs) for a given mass.
    
    Formula: Rs = 2GM / c^2
    
    Args:
        mass_kg (float): The mass of the body in kilograms.
        
    Returns:
        float: The Schwarzschild radius in meters.
    """
    # Physical constants
    G = 6.67430e-11  # Gravitational constant (m^3 kg^-1 s^-2)
    c = 299792458    # Speed of light (m/s)
    
    # Calculate Schwarzschild radius
    # Rs = (2 * G * M) / c^2
    rs = (2 * G * mass_kg) / (c**2)
    
    return rs

def main():
    # Example: Calculate the Schwarzschild radius for the Sun
    # Sun's mass is approximately 1.989 x 10^30 kg
    sun_mass = 1.989e30
    
    rs_sun = calculate_schwarzschild_radius(sun_mass)
    
    # Print the result using f-strings
    print(f"Object Mass: {sun_mass:.2e} kg")
    print(f"Calculated Schwarzschild Radius: {rs_sun:.4f} meters")

if __name__ == "__main__":
    main()
```

```python
def calculate_schwarzschild_radius(mass_kg):
    """
    Calculates the Schwarzschild radius (Rs) for a given mass.
    Formula: Rs = 2GM / c^2
    """
    # Physical constants
    G = 6.67430e-11  # m^3 kg^-1 s^-2
    c = 299792458    # m/s
    
    # Calculate Schwarzschild radius
    rs = (2 * G * mass_kg) / (c**2)
    return rs

def get_celestial_bodies():
    """
    Returns a dictionary of celestial bodies and their masses in kg.
    """
    sun_mass = 1.989e30
    return {
        "Earth": 5.972e24,
        "Sun": sun_mass,
        "Supermassive Black Hole": 1e8 * sun_mass
    }

def main():
    # Retrieve the dictionary of bodies
    bodies = get_celestial_bodies()
    
    print(f"{'Object':<25} | {'Mass (kg)':<15} | {'Schwarzschild Radius (m)'}")
    print("-" * 75)
    
    # Iterate through the dictionary to calculate and print results
    for name, mass in bodies.items():
        rs = calculate_schwarzschild_radius(mass)
        print(f"{name:<25} | {mass:<15.2e} | {rs:.4f} m")

if __name__ == "__main__":
    main()
```

```python
import math

def calculate_escape_velocity(mass_kg, radius_meters):
    """
    Calculates the escape velocity (ve) of a body.
    
    Formula: ve = sqrt(2GM / R)
    
    Args:
        mass_kg (float): The mass of the body in kilograms.
        radius_meters (float): The radius of the body in meters.
        
    Returns:
        float: The escape velocity in meters per second (m/s).
    """
    # Physical constants
    G = 6.67430e-11  # Gravitational constant (m^3 kg^-1 s^-2)
    
    # Calculate escape velocity using the provided formula
    # ve = sqrt((2 * G * M) / R)
    velocity = math.sqrt((2 * G * mass_kg) / radius_meters)
    
    return velocity

def main():
    # Example: Calculate the escape velocity for Earth
    # Earth mass ≈ 5.972 x 10^24 kg, Radius ≈ 6.371 x 10^6 meters
    earth_mass = 5.972e24
    earth_radius = 6.371e6
    
    ve_earth = calculate_escape_velocity(earth_mass, earth_radius)
    
    # Print the result using f-strings and convert to km/s for readability
    print(f"Earth Mass:   {earth_mass:.2e} kg")
    print(f"Earth Radius: {earth_radius:.2e} m")
    print(f"Escape Velocity: {ve_earth:.4f} m/s ({ve_earth/1000:.4f} km/s)")

if __name__ == "__main__":
    main()
```
