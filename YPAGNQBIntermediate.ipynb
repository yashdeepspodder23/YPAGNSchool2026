```python
import math

def calculate_stellar_luminosity(radius_meters, temperature_kelvin):
    """
    Computes stellar luminosity (L) using the Stefan-Boltzmann Law.
    
    Formula: L = 4 * pi * R^2 * sigma * T^4
    
    Args:
        radius_meters (float): Stellar radius in meters.
        temperature_kelvin (float): Effective surface temperature in Kelvin.
        
    Returns:
        float: Stellar luminosity in Watts (W).
    """
    # Stefan-Boltzmann constant (W m^-2 K^-4)
    sigma = 5.670374419e-8
    
    # Calculate luminosity
    # L = 4 * pi * R^2 * sigma * T^4
    luminosity = 4 * math.pi * (radius_meters**2) * sigma * (temperature_kelvin**4)
    
    return luminosity

def main():
    # Example: Calculate for the Sun
    # R_sun ≈ 6.9634e8 m, T_sun ≈ 5778 K
    sun_radius = 6.9634e8
    sun_temp = 5778
    
    lum = calculate_stellar_luminosity(sun_radius, sun_temp)
    
    print(f"Stellar Radius:   {sun_radius:.2e} m")
    print(f"Surface Temp:     {sun_temp} K")
    print(f"Calculated Luminosity: {lum:.4e} W")

if __name__ == "__main__":
    main()
```

```python
def classify_stellar_type(temp):
    """
    Classifies a star based on temperature using Morgan-Keenan boundaries.
    
    Returns:
        str: The spectral class (O, B, A, F, G, K, or M).
    """
    if temp >= 30000:
        return "O"
    elif temp >= 10000:
        return "B"
    elif temp >= 7500:
        return "A"
    elif temp >= 6000:
        return "F"
    elif temp >= 5200:
        return "G"
    elif temp >= 3700:
        return "K"
    else:
        return "M"

def main():
    # List of arbitrary stellar temperatures in Kelvin
    stellar_temperatures = [3500, 5800, 15000, 32000, 6500, 4500, 8000]
    
    # Table header
    print(f"{'Temperature (K)':<15} | {'Spectral Type'}")
    print("-" * 30)
    
    # Loop through temperatures and classify them
    for t in stellar_temperatures:
        spectral_type = classify_stellar_type(t)
        print(f"{t:<15} | {spectral_type}")

if __name__ == "__main__":
    main()
```

```python
def calculate_eddington_luminosity(mass_solar_units):
    """
    Calculates the Eddington luminosity limit (LE) for an object.
    
    Formula: LE = 1.26e38 * (M / M_sun) erg/s
    
    Args:
        mass_solar_units (float): The mass of the object in solar masses (M⊙).
        
    Returns:
        float: The Eddington luminosity in erg/s.
    """
    # Constant for Eddington limit (erg/s per solar mass)
    CONSTANT = 1.26e38
    
    # Calculate luminosity
    le = CONSTANT * mass_solar_units
    
    return le

def main():
    # Example: Calculate for a supermassive black hole of 10^8 M⊙
    mass_m_sun = 1e8
    
    luminosity = calculate_eddington_luminosity(mass_m_sun)
    
    print(f"Object Mass: {mass_m_sun:.1e} M⊙")
    print(f"Eddington Luminosity Limit: {luminosity:.4e} erg/s")

if __name__ == "__main__":
    main()
```

```python
import math

def calculate_distance_from_modulus(m, M):
    """
    Calculates the distance (d) in parsecs given apparent (m) 
    and absolute (M) magnitudes.
    
    Formula: d = 10 * 10^((m - M) / 5)
    """
    distance_pc = 10 * (10 ** ((m - M) / 5))
    return distance_pc

def main():
    # List of stars with (m, M) tuples
    # Format: (Name, Apparent Magnitude, Absolute Magnitude)
    star_data = [
        ("Sirius", -1.46, 1.43),
        ("Vega", 0.03, 0.58),
        ("Betelgeuse", 0.42, -5.85),
        ("Proxima Cen", 11.05, 15.53)
    ]
    
    print(f"{'Star':<15} | {'Distance (pc)':<15}")
    print("-" * 35)
    
    for name, m, M in star_data:
        d = calculate_distance_from_modulus(m, M)
        print(f"{name:<15} | {d:<15.4f}")

if __name__ == "__main__":
    main()
```
